```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Test.csv')
df
```

```python
df= df.drop(columns = ['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type'])
```

```python
df.info()
```

```python
X = df.drop(columns = 'Outlet_Type')
y = df['Outlet_Type']
```

```python
num_cols=X.select_dtypes(include='number').columns #X coz we've to choose from features
obj_cols=X.select_dtypes(include='object').columns
```

```python
X[obj_cols] = X[obj_cols].astype('category')
```

```python
df.info()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
numeric_columns = X.columns

preprocessor = ColumnTransformer(
    transformers=[
        ("scaler", StandardScaler(), numeric_columns)
    ]
)
pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", KNeighborsRegressor())
    ]
)
```

```python
pipeline.fit(xtrain, ytrain)
```
